```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('C:/Users/Admin/Desktop/tracks.csv')
df_tracks.head()
```

```python
# Null Values in my dataset
## checking the total null values present in my dataset

pd.isnull(df_tracks).sum()


```

```python
# To view the top 10 songs

sorted_df = df_tracks.sort_values('popularity', ascending = False).head(10)
sorted_df
```

```python
# converting the duration of the songs from ms to s

df_tracks['duration']= df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
# Putting release_date in place of the serial number.

df_tracks.set_index('release_date', inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks.duration.head()
```

```python
# Correlation between variables
corr_df=df_tracks.drop(['key','mode', 'explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidth=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Taking a sample data of 0.4% of my original data
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# Plot for energy vs Loudness
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = 'c', line_kws={"color": "red"}).set(title= "Loudness vs Energy Correlation")
```

```python
# Plot for energy vs Loudness
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = 'b', line_kws={"color": "red"}).set(title= "Popularity vs Acousticness Correlation")
```

```python
# Creating a new column 'year' from 'release_date' column

df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
# pip install --user seaborn==0.11.0 (This updates the sns)
```

```python
# Creating a distribution plot of the total songs

sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of Songs per year')
```

```python
# Song duration over the years

total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
# Average duration of songs over the years.

total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=30)
```
